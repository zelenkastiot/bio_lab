# Покушај

```python
from Bio import SeqIO

for seq_record in SeqIO.parse("sequence.gb", "genbank"):
    print(seq_record.id)
    print(repr(seq_record.seq))
    print(len(seq_record))
```

```python
print(seq_record.features[22])
```

```python
CDS_list = []

for i in range(0, len(seq_record.features)):
    if seq_record.features[i].type == "CDS":
        CDS_list.append(i)

print(f"Број на пронајдени CDS: {len(CDS_list)}")
```

```python
for i in CDS_list:
    print(seq_record.features[i])
```
